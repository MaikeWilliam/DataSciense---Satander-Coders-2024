```python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("USA_Housing.csv")

df.head()
```

```python
df.corr() #verificar a correlação
```

```python
x = df[["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Area Population"]] #Caracteristicas
y = df["Price"] #Alvo
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y) #Separamos em um grupo de treino e um grupo para ele testar
```

```python
modelo = LinearRegression() #Definimos o modelo
```

```python
modelo.fit(x_treino, y_treino) #Aqui ele aprende
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
y_previsto
```

```python
mean_absolute_error(y_previsto, y_teste)
```
