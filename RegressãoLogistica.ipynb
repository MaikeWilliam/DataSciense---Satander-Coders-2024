```python
import pandas as pd

from sklearn.linear_model import LogisticRegression #Classificador
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
df = pd.read_csv("data.csv")    
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')] # apaga coluna vazia
df
```

```python
x = df.drop(["id", "diagnosis"], axis=1)
```

```python
y = df["diagnosis"]
```

```python
modelo = LogisticRegression(max_iter=10000)
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
modelo.fit(x_treino, y_treino)
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
ConfusionMatrixDisplay.from_predictions(y_previsto, y_teste)
```

```python
print(classification_report(y_previsto, y_teste))
```
