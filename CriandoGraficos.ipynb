```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df.truncate(before=1, after=100)
```

```python
df = pd.read_csv("IMDB-Movie-Data.csv")
df = df.drop(df.columns[0], axis=1)

df_primeiras_20 = df.head(20)
```

```python
eixo_x = df_primeiras_20["Genre"].value_counts().keys()
eixo_y = df_primeiras_20["Genre"].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)

plt.title("Filmes por genero")
plt.xticks(rotation=45)

plt.xlabel("Generos de filme")
plt.ylabel("Quantidade de filmes")

plt.show()
```

```python
eixo_x = df_primeiras_20["Year"].value_counts().sort_index().keys()
eixo_y = df_primeiras_20["Year"].value_counts().sort_index().values
```

```python
plt.plot(eixo_x, eixo_y)

plt.title("Melhores filmes por ordem de lanamento")

plt.xlabel("Ano de lanamento")
plt.ylabel("Quantidade de filmes")

plt.show()
```
